```python
import numpy as np
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.plot(df['release_year'], df['movie_budget'], label='movie_Budget', color='blue')
plt.plot(df['release_year'], df['income_world'], label='Income World', color='red')
plt.title('Movie Budget and Income World over Time')
plt.xlabel('Release Year')
plt.ylabel('Amount in USD')
plt.legend(loc= "upper left")
plt.show()

#What are the trends of movie budget and income world along the time?
#The trends reveal significant insights into the movie industry. The graph shows how the financial aspects of filmaking have evolved over decades.
# Due to changes of production costs, audience demand, and market dynamics. 
#From the 1960's to the 2020's, movie budgets have increased overall. 
```

```python
year_data = df['release_year']
income_data = df['income_world']
movie_data = df['movie_budget']

plt.bar(year_data, income_data / movie_data, color= 'blue')
plt.title('Income to Budget Ratio over Time')
plt.xlabel('Release Year')
plt.ylabel('Income to Budget Ratio')
plt.show()
#What is the trend of income world – movie budget ratio along the time?
#The trend is that the income budget ration has fluctuated over the years. In the earlier decades the ratio was low.
#Which means that in the 1960's to the 1970's the films mainly didn't gain much income compared to their production costs.
#However in the 1980's to the 1990's there is a sudden increase in the ratio, indicating that the films began to earn more compared to their budgets. 

#From this trend, do you think producer make more profit now than they did in past?
#From 2010 to 2020, the income to budget ratio have changed to a higher level. 
#Which suggests that producers are making making more profit now than they did in the past, they are making lots of profit/revenue. 
```
